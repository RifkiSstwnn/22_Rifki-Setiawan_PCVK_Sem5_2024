```python
from typing import final
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from PIL import Image
import matplotlib.pylab as plt

urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://s0.bukalapak.com/uploads/content_attachment/a00e2fe630e8d7624b0135c5/original/Ikan_Cupang.jpg"]

# baca dan tampilkan image berapa dapat dari list urls
for url in urls:
    # load url to opencv
    image = io.imread(url)
    image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
    image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
    final_frame = cv.hconcat((image, image_2))
    cv2_imshow(final_frame)
    print('\n')


tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("Resolusi image: tinggi x lebar = ", tinggi, "x", lebar)
cv2_imshow(image_2)

image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal di tengah gambar
for y in range (lebar):
  image_3[int((tinggi)/2), y] = [255, 255, 255]

#membuat garis vertikal di tengah gambar
for x in range (tinggi):
  image_4[x, lebar//2] = [255, 255, 255]


final_frame1 = cv.hconcat((image_2, image_3))
final_frame2 = cv.hconcat((image_2, image_4))
final_frame = cv.vconcat((final_frame1, final_frame2))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Membuat garis silang diagonal pada gambar
for x in range(tinggi):
    y1 = int((lebar / tinggi) * x)
    y2 = lebar - y1
    if 0 <= y1 < lebar:
        image_3[x, y1] = [255, 255, 255]
    if 0 <= y2 < lebar:
        image_3[x, y2] = [255, 255, 255]

# Menggabungkan gambar asli dengan garis silang
final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
